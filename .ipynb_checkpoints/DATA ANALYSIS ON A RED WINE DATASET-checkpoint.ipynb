```python
class RedWineDataset:
    def __init__(self, red_wine_data):
        self.red_wine_data = red_wine_data
        
    def head_preview(self, n=10):
            return self.red_wine_data.head(n)
        
    def tail_preview(self, n=20):
        return self.red_wine_data.tail(n)

    def info_preview(self):
        return self.red_wine_data.info()

    def describe_preview(self):
        return self.red_wine_data.describe()

red_wine_data = pd.read_csv('red_wine.csv')

red_wine = RedWineDataset(red_wine_data)

#previewing the head of the data
head_data = red_wine.head_preview()
print("Head of the red wine dataset:")
print(head_data)

#previewing the tail of the data
tail_data = red_wine.tail_preview()
print("\nTail of the red wine dataset:")
print(tail_data)

#previewing the information of the data
info_data = red_wine.info_preview()
print("\nInformation about the red wine dataset:")
print(info_data)

#previewing descriptive statistical information of the data using the describe method 
describe_data = red_wine.describe_preview()
print("\nDescriptive statistics of the red wine dataset:")
print(describe_data)
```

```python
#Graph 1
#Plotting graph for the data
modified_wine_data= pd.read_csv('red_wine_modified.csv')
plt.figure(figsize =(8,10))
sns.histplot(modified_wine_data['residual sugar'], bins=15, kde=True, label='residual sugar', color='b', edgecolor='y')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('GRAPH OF Residual Sugar VALUES')
plt.show()
```

# REPORT ON THE RED WINE DATASET


 
### In this red wine dataset the main features were, *fixed acidity, volatile acidity, citric acid, residual sugar, chlorides, free sulfur dioxide, total sulfur dioxide, density, pH, sulphates, alcohol, quality*. All these features play an important role im making sure red wine has the right taste, colour, aroma, and composition. These key features will make it possible to make a sound analysis on the red wine dataset which will help consumers to choose red wines suitable for their taste. In this report, analysis was done on the red wine, and graphs were drawn on four features to help make the analysis more practical and insightful. The first graph(graph 1) was about the residual sugar of the red wine. Residual sugar makes the chemical composition of the wine, it determines how sweet the wine will be. Most of the time red wine is preferred to have less sugar in it, so the sugar is mostly fermented to alcohol. In this analysis it is seen that the amount of residual sugar in this red wine is quite high, hence it helps a winemaker or consumer to group it as a sweet wine. It will also help winemakers to control the amount of residual sugar in wines to prevent them from not achieving their purpose

### The second graph represents the pH values of the wine, pH contribute to the sourness or bitterness of the wine, a lower pH is necessary for red wine, the graph shows that the wine has more pH values ranging from 3.2 to 3.4, which is a good value for a wine to taste okay.

### The third graph was a correlation matrice of all the variables of the red wine. Comparison of how the various features affect the quality of the wine. Where there exists a positive correlation it means that the variable has a good effect on the quality of the wine, but vice versa for a negative correlation. The colors also help to identify a positive and negative correlation that exists between the features. Only three variables show strong correlation, they are sulphates, alcohol, and quality itself. This means a reasonable amount of sulphate and alcohol content may influence the quality rating of wine positively. Sulphate is used as a preservative for red wine. Those that have weak and negative correlation are needed in a small amount to make the wine taste better.

### Winemakers and wine enthusiasts can use this analysis of the dataset to check the quality rating of wine and improve upon the taste and othe chemical properties
